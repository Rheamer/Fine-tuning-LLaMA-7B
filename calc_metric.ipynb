```
import matplotlib.pyplot as plt

n = 65 # число подкатов
n_gpt = 17 # число gpt
n_llama = 48 # число llama

data_list = [n_llama, n_gpt]
labels_list = ['LLaMA-7B', 'Chat-GPT']

data = list(data_list)
labels = list(labels_list)
plt.pie(data, labels=labels, autopct='%1.1f%%')
plt.show()
```

```
list_answer = []
for i in range(n):
    inp = str(input()).split()
    for i in range(len(inp)):
        inp[i] = int(inp[i])
    list_answer.append(inp)
```

```
list_gpt = [3, 7, 11, 16, 20, 24, 28, 32, 36, 41, 44, 46, 52, 55, 60, 62, 64]
```

```
import numpy as np
list_gpt_index = np.array([0, 0, 0]) #1,2,3
list_llama_index = np.array([0, 0, 0]) #1,2,3
list_diff = np.array([0, 0, 0])
all_gpt = 0
all_llama = 0
for i in range(n):
    list_diff = list_diff + np.array(list_answer[i])
    if i in list_gpt:
        for j in range(3):
            all_gpt = all_gpt + list_answer[i][j]
        list_gpt_index = list_gpt_index + np.array(list_answer[i])
    else:
        for j in range(3):
            all_llama = all_llama + list_answer[i][j]
        list_llama_index = list_llama_index + np.array(list_answer[i])
```

```
n_llama*list_gpt_index/list_diff/n, n_gpt*list_llama_index/list_diff/n
```

```
list_gpt_index/all_gpt, list_llama_index/all_llama
```

```
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))
labels = ['Не похоже на подкат','Похоже на подкат','Хороший подкат']
names = ['LLaMA-7B', 'Chat-GPT']

ax[0].pie([n_gpt*list_llama_index[0]/list_diff[0], n_llama*list_gpt_index[0]/list_diff[0]], labels=names, autopct='%1.1f%%')
ax[0].set_title(labels[0])

ax[1].pie([n_gpt*list_llama_index[1]/list_diff[1], n_llama*list_gpt_index[1]/list_diff[1]], labels=names, autopct='%1.1f%%')
ax[1].set_title(labels[1])

ax[2].pie([n_gpt*list_llama_index[2]/list_diff[2], n_llama*list_gpt_index[2]/list_diff[2]], labels=names, autopct='%1.1f%%')
ax[2].set_title(labels[2])

plt.legend()
plt.show()
```

```
plt.pie(list_llama_index/all_llama, labels=['Не похоже на подкат','Похоже на подкат','Хороший подкат'], autopct='%1.1f%%')
plt.title('LLaMA-7B')
plt.show()
```
